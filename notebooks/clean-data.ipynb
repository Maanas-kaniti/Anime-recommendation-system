```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('data/anime_planet_full_details.csv')
data.head()

data.info()
```

```python
data["type_only"] = data["type"].astype(str).str.split("(").str[0].str.strip()
data["episodes"] = data["type"].astype(str).str.extract(r"\((.*?)\)")
data.to_csv('data/anime_planet_full_details_.csv',index= False)
```

```python
df = pd.read_csv('data/anime_planet_full_details_.csv')
df['type_only'].unique()


df
```

```python
df['episodes'].fillna('Unknown',inplace=True)
df['episodes'] = df['episodes'].astype(str).str.split(" ").str[0]
df['episodes'] = df['episodes'].str.replace('+', '', regex=False)
df['episodes'] = df['episodes'].replace('Unknown',np.nan)
df['episodes'] = pd.to_numeric(df['episodes'], errors='coerce')
```

```python
df['genres'].fillna('[]',inplace = True)
(df['genres']=='[]').sum()
```

```python
df.to_csv('data/anime_planet_full_details_cleaned.csv',index = False)
```
